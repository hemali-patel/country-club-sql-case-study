```python
# Q2
query = """
SELECT COUNT(*)
FROM Facilities
WHERE membercost = 0;
"""
pd.read_sql_query(query, conn)
```

```python
# Q3
query = """
SELECT facid, name, membercost, monthlymaintenance
FROM Facilities
WHERE membercost > 0
AND membercost < 0.2 * monthlymaintenance;
"""

pd.read_sql_query(query, conn)
```

```python
# Q4
query = """
SELECT *
FROM Facilities
WHERE facid IN (1, 5);
"""

pd.read_sql_query(query, conn)
```

```python
# Q5
query = """
SELECT name,
       monthlymaintenance,
       CASE 
           WHEN monthlymaintenance > 100 THEN 'expensive'
           ELSE 'cheap'
       END AS cost_label
FROM Facilities;
"""

pd.read_sql_query(query, conn)
```

```python
# Q6
query = """
SELECT firstname, surname
FROM Members
WHERE joindate = (SELECT MAX(joindate) FROM Members);
"""

pd.read_sql_query(query, conn)
```

```python
# Q7
query = """
SELECT DISTINCT
    f.name AS court_name,
    m.firstname || ' ' || m.surname AS member_name
FROM Bookings b
JOIN Facilities f ON b.facid = f.facid
JOIN Members m ON b.memid = m.memid
WHERE f.name LIKE 'Tennis Court%'
ORDER BY member_name;
"""

pd.read_sql_query(query, conn)
```

```python
# Q8
query = """
SELECT 
    f.name AS facility,
    m.firstname || ' ' || m.surname AS member,
    CASE 
        WHEN b.memid = 0 THEN b.slots * f.guestcost
        ELSE b.slots * f.membercost
    END AS cost
FROM Bookings b
JOIN Facilities f ON b.facid = f.facid
JOIN Members m ON b.memid = m.memid
WHERE DATE(b.starttime) = '2012-09-14'
AND (
    CASE 
        WHEN b.memid = 0 THEN b.slots * f.guestcost
        ELSE b.slots * f.membercost
    END
) > 30
ORDER BY cost DESC;
"""

pd.read_sql_query(query, conn)
```

```python
# Q9
query = """
SELECT *
FROM (
    SELECT 
        f.name AS facility,
        m.firstname || ' ' || m.surname AS member,
        CASE 
            WHEN b.memid = 0 THEN b.slots * f.guestcost
            ELSE b.slots * f.membercost
        END AS cost
    FROM Bookings b
    JOIN Facilities f ON b.facid = f.facid
    JOIN Members m ON b.memid = m.memid
    WHERE DATE(b.starttime) = '2012-09-14'
) sub
WHERE cost > 30
ORDER BY cost DESC;
"""

pd.read_sql_query(query, conn)
```

```python
# Q10
query = """
SELECT f.name,
       SUM(
           CASE 
               WHEN b.memid = 0 THEN b.slots * f.guestcost
               ELSE b.slots * f.membercost
           END
       ) AS revenue
FROM Bookings b
JOIN Facilities f ON b.facid = f.facid
GROUP BY f.name
HAVING revenue < 1000
ORDER BY revenue;
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# Q11
query = """
SELECT 
    m.surname || ', ' || m.firstname AS member_name,
    r.surname || ', ' || r.firstname AS recommender_name
FROM Members m
LEFT JOIN Members r 
    ON m.recommendedby = r.memid
ORDER BY m.surname, m.firstname;
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# Q12
query = """
SELECT 
    f.name AS facility,
    COUNT(b.bookid) AS member_usage
FROM Bookings b
JOIN Facilities f ON b.facid = f.facid
WHERE b.memid != 0
GROUP BY f.name
ORDER BY member_usage DESC;
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# Q13
query = """
SELECT 
    f.name AS facility,
    strftime('%m', b.starttime) AS month,
    COUNT(b.bookid) AS usage
FROM Bookings b
JOIN Facilities f ON b.facid = f.facid
WHERE b.memid != 0
GROUP BY f.name, month
ORDER BY f.name, month;
"""

df = pd.read_sql_query(query, conn)
df
```
